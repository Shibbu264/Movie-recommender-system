```python


import numpy as np
import pandas as pd
```

```python
movies =pd.read_csv('tmdb_5000_movies[1].csv')
credits =pd.read_csv('tmdb_5000_credits[1].csv')
```

```python
movies.head(5)
```

```python
credits.head(1)
```

```python
movies=movies.merge(credits,on='title')
movies.head(1)
```

```python
# genre id keywords title overview cast crew
```

```python
movies=movies[['movie_id','title','cast','crew','overview','genres','keywords']]
movies.head(1)
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
movies=movies.drop_duplicates()
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
 L = []
 for i in ast.literal_eval(obj):
  L.append(i['name'])
 return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert3(obj):
 L = []
 counter = 0
 for i in ast.literal_eval(obj):
  if counter != 3:
   L.append(i['name'])
   counter+=1
  else:
   break
 return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_direc(obj):
 L = []
 for i in ast.literal_eval(obj):
     if i['job']== 'Director':
          L.append(i['name'])
         
     
   
 return L
```

```python
movies['title']
```

```python
movies['crew']=movies['crew'].apply(fetch_direc)
```

```python
movies.head()
```

```python
movies ['overview']= movies ['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['cast']+movies['crew']+movies['keywords']+movies['genres']
```

```python
movies['tags']=movies['tags']+movies ['overview']
```

```python
movies.head(1)
```

```python
new_df=movies[['movie_id','title','tags']]
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
        
```

```python
new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity =cosine_similarity(vectors)
```
